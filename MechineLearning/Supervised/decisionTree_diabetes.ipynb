```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score

# Membaca file iris.csv
df_diabetes = pd.read_csv('../../data_samples/diabetes.csv')


df_diabetes['Outcome'] = df_diabetes['Outcome'].where(df_diabetes['Outcome'] != 0, 'free')
df_diabetes['Outcome'] = df_diabetes['Outcome'].where(df_diabetes['Outcome'] != 1, 'diabetes')

df_diabetes.info()
df_diabetes[20:89]
```

```python
# Memisahkan atribut dan label
df_attr = df_diabetes[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
df_label = df_diabetes['Outcome']

# print(df_attr)
# print(df_label)

# Membagi dataset menjadi data train dan data test
attr_train, attr_test, label_train, label_test = train_test_split(df_attr, df_label, test_size=0.2, random_state=1)
```

```python
# membuat model tree
tree_model = DecisionTreeClassifier()

# Training dataset kedalam model tree
tree_model = tree_model.fit(attr_train, label_train)

# Evaluasi Model Tree
label_pred = tree_model.predict(attr_test)

acc_score = round(accuracy_score(label_pred, label_test), 3)
print("tingkat akurasi: ", acc_score)
```
