```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.metrics import accuracy_score

# Membaca file iris.csv
df_iris = pd.read_csv('../../data_samples/Iris.csv')


df_iris.info()
df_iris[20:89]
```

```python
# Data cleaning
df_iris.drop('Id', axis=1, inplace=True)


# Memisahkan atribut dan label
df_attr = df_iris[['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm', 'PetalWidthCm']]
df_label = df_iris['Species']

# print(df_attr)
# print(df_label)

# Membagi dataset menjadi data train dan data test
attr_train, attr_test, label_train, label_test = train_test_split(df_attr, df_label, test_size=0.2, random_state=123)
```

```python
# membuat model tree
tree_model = DecisionTreeClassifier()

# Training dataset kedalam model tree
tree_model = tree_model.fit(attr_train, label_train)

# Evaluasi Model Tree
label_pred = tree_model.predict(attr_test)

acc_score = round(accuracy_score(label_pred, label_test), 3)
print("tingkat akurasi: ", acc_score)
```

```python
# prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
obj_one = [3, 5.0, 3.7, 4.5]
print(obj_one)

obj_one_pred = tree_model.predict([[obj_one]][0])

print(obj_one_pred)
```

```python
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ],
    rounded= True,
    filled =True
)
```
